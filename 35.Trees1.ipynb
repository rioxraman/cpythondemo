# BINARY TREE HAVING AT MOST 2 CHILD

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

n1 = Node(10)
n2 = Node(20)
n3 = Node(15)
n1.left = n2
n1.right = n3
print(n1)
```

```python
n1.left.data = 30
print(n1)
n1.left.left = Node(23)
print(n1)
```

```python
def tranversal(node):
    if node is None:
        return None
    tranversal(node.left)
    tranversal(node.right)
```

```python
# given root of binary trree, find size (count of node)
def size(node):
    if node is None:
        return 0

    return 1+size(node.left)+size(node.right)
```

```python
# given root of binary trree, find sum (sum of nodes)
def sum(node):
    if node is None:
        return 0

    return sum(node.left)+sum(node.right)+node.data
sum(n1)
```
