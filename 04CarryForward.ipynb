```python
class Solution(object):
    def replaceElements(self, arr):
        """
        :type arr: List[int]
        :rtype: List[int]
        """
        max_so_far = -1
        n=len(arr)
        for i in range(-1,-(n+1),-1):
            arr[i],max_so_far=max_so_far,max(max_so_far,arr[i])
        return arr
```

```python
class Solution(object):
    def replaceElements(self, arr):

        mx = -1

        for i in range(len(arr)-1, -1, -1):

            temp = arr[i]
            arr[i] = mx

            if temp > mx:
                mx = temp

        return arr
```
