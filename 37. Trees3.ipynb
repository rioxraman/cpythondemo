Introduction to BSTs:
A BST is a binary tree where each node adheres to specific ordering properties: nodes in the left subtree have values smaller than the current node, and nodes in the right subtree have values greater than the current node.
This property must hold true for every node within the tree.


Searching in a BST:
BSTs enable efficient searching by allowing the algorithm to discard either the left or right subtree based on the comparison of the target value against the current node's value.
The time complexity for searching is proportional to the height of the tree ($O(H)$).


Inserting into a BST:
Insertion involves finding the correct position for a new value $K$ by traversing the tree similarly to a search operation.
When the traversal reaches a null (or None) position, a new node is created and linked to the parent.
Existing tree structure remains intact, and duplicates are typically not inserted.


BST Verification:
A BST can be verified by performing an in-order traversal, which must result in a sorted sequence of values.
An efficient way to check this is to perform the traversal and compare the current node's value with the previously visited node's value; if the previous value is ever greater than or equal to the current, the tree is not a valid BST.


Creating a Balanced BST:
To construct a balanced BST from a sorted array, the middle element is selected as the root node to maintain balance.
The algorithm then recursively applies this logic to the left subarray (to form the left subtree) and the right subarray (to form the right subtree).
