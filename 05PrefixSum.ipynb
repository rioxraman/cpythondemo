```python
arr = [-3,6,2,4,5,2,8,-9,3,1]

n = len(arr)

prefix = [0] * n

prefix[0] = arr[0]

# build prefix sum
for i in range(1, n):
    prefix[i] = prefix[i-1] + arr[i]

print(prefix)


def range_sum(arr, l, r):

    if l == 0:
        return prefix[r]

    return prefix[r] - prefix[l-1]


print(range_sum(arr, 2, 5))
```
